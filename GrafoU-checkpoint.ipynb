# Grafo de la Universidad

## Importar las librerías

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Función para calcular la ruta más corta entre dos puntos en un grafo

```python
def calcular_ruta_mas_corta(grafo, inicio, fin):
    try:
        # Utilizamos el algoritmo de Dijkstra para calcular la ruta más corta
        ruta = nx.dijkstra_path(grafo, inicio, fin)
        distancia = nx.dijkstra_path_length(grafo, inicio, fin)
        return ruta, distancia
    except nx.NetworkXNoPath:
        return None, None
```

## Función para visualizar el grafo y la ruta más corta

```python
def visualizar_grafo_y_ruta(grafo, ruta, inicio, fin):
    posiciones = nx.spring_layout(grafo)  # Obtener posiciones de los nodos para la visualización
    nx.draw(grafo, pos=posiciones, with_labels=True, node_size=700, node_color='lightblue', font_size=10)  # Dibujar el grafo
    if ruta:
        # Dibujar la ruta más corta
        edge_labels = {(ruta[i], ruta[i + 1]): grafo.get_edge_data(ruta[i], ruta[i + 1])['weight'] for i in range(len(ruta) - 1)}
        nx.draw_networkx_edge_labels(grafo, pos=posiciones, edge_labels=edge_labels, font_color='red')
        nx.draw_networkx_edges(grafo, pos=posiciones, edgelist=[(ruta[i], ruta[i + 1]) for i in range(len(ruta) - 1)], edge_color='red', width=2)
        print("La ruta más corta entre", inicio, "y", fin, "es:", ruta)
        print("La distancia de la ruta más corta es:", nx.dijkstra_path_length(grafo, inicio, fin))
    else:
        print("No hay ruta entre", inicio, "y", fin)
    plt.title("Grafo y Ruta más Corta")
    plt.show()
```

## Crear el grafo

```python
grafo = nx.Graph()

#Edificios
grafo.add_edge('AD PORTAS', 'K', weight=1)
grafo.add_edge('K', 'A', weight=1)
grafo.add_edge('A', 'B', weight=1)
grafo.add_edge('B', 'C', weight=1)
grafo.add_edge('C', 'A', weight=1)
grafo.add_edge('B', 'E', weight=6)

#Cafeterías y restaurantes
```

## Solicitar puntos de inicio y destino por consola

```python
inicio = input("Ingrese el punto de inicio: ")
fin = input("Ingrese el punto de destino: ")
```

## Calcular y visualizar la ruta más corta

```python
ruta, _ = calcular_ruta_mas_corta(grafo, inicio, fin)
visualizar_grafo_y_ruta(grafo, ruta, inicio, fin)
```
