## Clase grafo

```python
class Grafo:
    def __init__(self):
        self.vertices = {}

    def agregar_vertice(self, vertice):
        if vertice not in self.vertices:
            self.vertices[vertice] = []

    def agregar_borde(self, origen, destino):
        self.agregar_vertice(origen)
        self.agregar_vertice(destino)
        self.vertices[origen].append(destino)

    def dfs(self, inicio, visitados=None):
        if visitados is None:
            visitados = set()
        visitados.add(inicio)
        print(inicio, end=' ')
        for vecino in self.vertices[inicio]:
            if vecino not in visitados:
                self.dfs(vecino, visitados)

    def bfs(self, inicio):
        visitados = set()
        cola = [inicio]
        while cola:
            vertice = cola.pop(0)
            if vertice not in visitados:
                print(vertice, end=' ')
                visitados.add(vertice)
                for vecino in self.vertices[vertice]:
                    cola.append(vecino)

# Creación de un objeto Grafo
mi_grafo = Grafo()
```

## Función para calcular la ruta más corta entre dos puntos en un grafo

```python
def calcular_ruta_mas_corta(grafo, inicio, fin):
    try:
        # Utilizamos el algoritmo de Dijkstra para calcular la ruta más corta
        ruta = nx.dijkstra_path(grafo, inicio, fin)
        distancia = nx.dijkstra_path_length(grafo, inicio, fin)
        return ruta, distancia
    except nx.NetworkXNoPath:
        return None, None
```

## Función para visualizar el grafo y la ruta más corta

```python
def visualizar_grafo_y_ruta(grafo, ruta, inicio, fin):
    posiciones = nx.spring_layout(grafo)  # Obtener posiciones de los nodos para la visualización
    nx.draw(grafo, pos=posiciones, with_labels=True, node_size=1000, node_color='lightgreen', font_size=10)  # Dibujar el grafo
    if ruta:
        # Dibujar la ruta más corta
        edge_labels = {(ruta[i], ruta[i + 1]): grafo.get_edge_data(ruta[i], ruta[i + 1])['weight'] for i in range(len(ruta) - 1)}
        nx.draw_networkx_edge_labels(grafo, pos=posiciones, edge_labels=edge_labels, font_color='red')
        nx.draw_networkx_edges(grafo, pos=posiciones, edgelist=[(ruta[i], ruta[i + 1]) for i in range(len(ruta) - 1)], edge_color='blue', width=2)
        print("La ruta más corta entre", inicio, "y", fin, "es:", ruta)
        print("La distancia de la ruta más corta es:", nx.dijkstra_path_length(grafo, inicio, fin))
    else:
        print("No hay ruta entre", inicio, "y", fin)
    plt.title("Grafo y Ruta más Corta")
    plt.show()
    
```
